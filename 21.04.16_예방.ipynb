```python
import csv
f = open('210416.csv',encoding = 'cp949')
data = csv.reader(f)
header = next(data)
s = 0
k = 0
d = 0
o = 0 
c = 0
c_1 = 0
r = 0

for i in data:
    
    if "서울" in i[6]:  
        s += 1
       
    elif "경기도" in i[6] or "인천" in i[6] or "경기" in i[6]: 
        k += 1
       
    elif "대구" in i[6] or "경상" in i[6]  or "울산" in i[6] or "부산" in i[6] or "경남" in i[6] or "경북" in i[6]:
        d += 1
        
    elif "대전" in i[6] or "충청" in i[6] or "세종" in i[6] or "충남" in i[6] or "충북" in i[6]:
        o += 1
        
    elif "전라" in i[6] or "광주" in i[6] or "전남" in i[6] or "전북" in i[6]:
        c += 1
        
    elif "강원도" in i[6]:
        c_1 += 1
        
    else :
        r += 1
        
print("서울:%d, 경기:%d, 경상도:%d, 충청도:%d, 전라도:%d, 강원도:%d, 제주:%d"%(s,k,d,o,c,c_1,r))
print("총개수: %d"%int(s+d+o+c+c_1+r+k))

f.close()

import matplotlib.pyplot as plt
#plt.rc('font',family = 'NanumGothic')
plt.figure(figsize = (10,10))
size = [s,k,d,o,c,c_1,r]
label = ['Seoul','Gyeonggi','Gyeongsang','Chungcheong','Jeolla', 'Gangwon','Jeju']
#plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%',textprops={'fontsize':13},colors = ['lightgray','aqua','chartreuse','tomato','mediumturquoise','lightyellow','lightpink'])
plt.legend()
plt.show()

    
```
