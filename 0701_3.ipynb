```python
import requests
from bs4 import BeautifulSoup


def top ():
    url = "https://www.op.gg/champion/statistics"
    htmlText = requests.get(url).text
    bsoup = BeautifulSoup(htmlText, "html.parser")
    btab = bsoup.find("tbody", {"class" : "tabItem champion-trend-tier-TOP"})
    btrs = btab.find_all("td",{"class" :"champion-index-table__cell champion-index-table__cell--champion"})
    
    tops = []

    for tr in btrs:
        dic = {}
        tds = tr.find_all("div")
        dic['champion'] = tds[0].text
        a = tds[1].text.replace("\t","")
        dic['position'] = a.replace("\n","")
        tops.append(dic)
        
def jungle ():
    url = "https://www.op.gg/champion/statistics"
    htmlText = requests.get(url).text
    bsoup = BeautifulSoup(htmlText, "html.parser")
    btab = bsoup.find("tbody", {"class" : "tabItem champion-trend-tier-JUNGLE"})
    btrs = btab.find_all("td",{"class" :"champion-index-table__cell champion-index-table__cell--champion"})
    
    jungles = []

    for tr in btrs:
        dic = {}
        tds = tr.find_all("div")
        dic['champion'] = tds[0].text
        a = tds[1].text.replace("\t","")
        dic['position'] = a.replace("\n","")
        jungles.append(dic)

def mid ():
    url = "https://www.op.gg/champion/statistics"
    htmlText = requests.get(url).text
    bsoup = BeautifulSoup(htmlText, "html.parser")
    btab = bsoup.find("tbody", {"class" : "tabItem champion-trend-tier-MID"})
    btrs = btab.find_all("td",{"class" :"champion-index-table__cell champion-index-table__cell--champion"})
    
    mids = []

    for tr in btrs:
        dic = {}
        tds = tr.find_all("div")
        dic['champion'] = tds[0].text
        a = tds[1].text.replace("\t","")
        dic['position'] = a.replace("\n","")
        mids.append(dic)
    
def adc ():
    url = "https://www.op.gg/champion/statistics"
    htmlText = requests.get(url).text
    bsoup = BeautifulSoup(htmlText, "html.parser")
    btab = bsoup.find("tbody", {"class" : "tabItem champion-trend-tier-ADC"})
    btrs = btab.find_all("td",{"class" :"champion-index-table__cell champion-index-table__cell--champion"})
    
    adcs = []

    for tr in btrs:
        dic = {}
        tds = tr.find_all("div")
        dic['champion'] = tds[0].text
        a = tds[1].text.replace("\t","")
        dic['position'] = a.replace("\n","")
        adcs.append(dic) 
    
def support ():
    url = "https://www.op.gg/champion/statistics"
    htmlText = requests.get(url).text
    bsoup = BeautifulSoup(htmlText, "html.parser")
    btab = bsoup.find("tbody", {"class" : "tabItem champion-trend-tier-SUPPORT"})
    btrs = btab.find_all("td",{"class" :"champion-index-table__cell champion-index-table__cell--champion"})
    
    supports = []
    
    for tr in btrs:
        dic = {}
        tds = tr.find_all("div")
        dic['champion'] = tds[0].text
        a = tds[1].text.replace("\t","")
        dic['position'] = a.replace("\n","")
        supports.append(dic) 
```

```python
from tkinter import Tk,ttk,Label,Button,Text,END
import tkinter as tk #tkinter 라이브러리

window = Tk()
window.title("챔피언 분석")  # 이름
window.geometry("500x500") # 크기 조정
window.resizable(0,0) #고정
        
        
title = "포지션별 챔피언 랭킹" #헤더 느낌?
title_feature = Label(window,text = title, font = ("Noto Sans KR",20)) #라벨 , 폰트 조정
title_feature.pack(padx = 10, pady = 15) 

treeTable = ttk.Treeview(window)  #테이블 생성
treeTable["columns"] = ("champion","position")  # 테이블 아이템
treeTable.column("#0",width = 50)  # 이름 및 테이블 크기
treeTable.column("champion",width = 100)  
treeTable.column("position",width = 50)

treeTable.heading("#0", text = "No.",anchor = "center") #heading열이름 / 중앙 정렬
treeTable.heading("champion", text = "챔피언", anchor = "center")  #테이블 아이템 이름 변경 / 중앙 정렬
treeTable.heading("position",text = "라인", anchor = "center")

OptionList = [
    "탑",
    "정글",
    "미드",
    "원딜",
    "서폿",
]

variable = tk.StringVar()
# variable.set(OptionList[0])

opt = ttk.Combobox(window, value = OptionList, state = "readonly" ,textvariable = variable)
opt.config(width=15, font=('Noto Sans KR', 12))
opt.place(x = 20, y = 50)


def setTableItem():
    treeTable.delete(*treeTable.get_children())
    country = opt.get()
#     for op in OptionList :
    if country== "탑":
        for idx, report in enumerate(top): #enumerate tops안에 리스트를 숫자화?
            champion = report['champion']
            position = report['position']
            treeTable.insert("",'end',iid = None, text = str(idx),values = [champion,position])
    elif country == "정글":
         for idx, report in enumerate(jungle): 
            champion = report['champion']
            position = report['position']
            treeTable.insert("",'end',iid = None, text = str(idx),values = [champion,position])
    elif country == "미드":
         for idx, report in enumerate(mid):
            champion = report['champion']
            position = report['position']
            treeTable.insert("",'end',iid = None, text = str(idx),values = [champion,position])  
    elif country == "원딜":
         for idx, report in enumerate(adc): 
            champion = report['champion']
            position = report['position']
            treeTable.insert("",'end',iid = None, text = str(idx),values = [champion,position])
    elif country == "서폿":
        for idx, report in enumerate(support): 
            champion = report['champion']
            position = report['position']
            treeTable.insert("",'end',iid = None, text = str(idx),values = [champion,position])

            
opt.btn = Button(window, width=10,text ='click',command=setTableItem)
opt.btn.pack()

# def click_item(event): 
#     selectedItem = treeTable.focus() 
#     getValue = treeTable.item(selectedItem).get('values') # 딕셔너리의 값만 가져오기 label.configure(text=getValue) # 라벨 내용 바꾸기
#     treeTable.configure(text=getValue)     
    
# btn1 = Button(window, text = '검색', command= click_item) #command = 함수이름
# btn1.place(x=-100,y=300, relx = 0.5, width = 40, height = 40)
# btn1.pack()

setTableItem()       
treeTable.place(x=10,y=90,width=380,height=200)

#conda install pyinstaller 
# cd <= 경로 지정 (cd C:\)
#pyinstaller -w -F 0701_2.py


window.mainloop()
```
